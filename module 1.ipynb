```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt




file_path = "dataaa.csv.csv"  
df = pd.read_csv(file_path)


df_filtered = df[['Year', 'Cyclonic Disturbances - TOTAL']].dropna()


sample_size = 40
sample = df_filtered.sample(n=sample_size, random_state=42)


sample_mean = sample['Cyclonic Disturbances - TOTAL'].mean()
sample_std = sample['Cyclonic Disturbances - TOTAL'].std(ddof=1)


population_mean = df_filtered['Cyclonic Disturbances - TOTAL'].mean()


z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))


z_critical = 1.645


reject_null = z_score > z_critical


plt.figure(figsize=(10, 5))
plt.hist(df_filtered['Cyclonic Disturbances - TOTAL'], bins=20, alpha=0.6, label="All Data")
plt.axvline(sample_mean, color='r', linestyle='dashed', linewidth=2, label=f'Sample Mean = {sample_mean:.2f}')
plt.axvline(population_mean, color='g', linestyle='dashed', linewidth=2, label=f'Population Mean = {population_mean:.2f}')
plt.legend()
plt.xlabel("Cyclonic Disturbances - TOTAL")
plt.ylabel("Frequency")
plt.title("Cyclonic Disturbances Distribution & Hypothesis Testing")
plt.show()


print(f"Z-score: {z_score:.3f}")
print(f"Reject Null Hypothesis: {reject_null}")
```
